### Importing the Required Libraries

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

### Importing the Data

```python
df=pd.read_csv('creditcard.csv')
```

### Exploratory Data Analysis

```python
df.describe
```

```python
df.info()
```

```python
corr=df.corr()
```

```python
plt.figure(figsize=(26,12))
sns.heatmap(corr,cmap='viridis',annot=True)
```

### Taking a Sample of the Data

```python
data=df.sample(frac=0.25,random_state=1)
```

```python
data.info()
```

### Checking the Spread of the Data 

```python
data.hist(figsize=(20,20))
```

```python
data['Class'].value_counts()
```

### Dropping the column as the values mentioned in the column are inconclusive 

```python
X=(data.drop('Class',axis=1))
y= data['Class']
```

### Applying Logistic Regression to generate the Model.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

### Splitting the Data into Train and Test Set. 

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30)
```

```python
model=LogisticRegression()
```

### Training the Model

```python
model.fit(X_train,y_train)
```

### Checking the quality of the results generated by the Algorithm

```python
from sklearn.metrics import classification_report
```

### Accuracy close to 99.88 for the given sample of data

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
prediction=model.predict(X_test)
Log_Reg=classification_report(y_test,prediction)
```
